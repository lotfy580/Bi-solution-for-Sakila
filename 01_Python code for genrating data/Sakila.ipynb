```python


mapping = {'Kazakstan':'Kazakhstan', 'Runion':'Madagascar', 'Yugoslavia':'Serbia', 'Holy See (Vatican City State)':'Italy'}
country['country'] = country['country'].replace(mapping)


def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
country['continent'] = 1
for i in range(len(country)):
    country.at[i, 'continent'] = country_to_continent(country.at[i, 'country'])
    
conditions = [
    (country['continent'] == 'Asia'),
    (country['continent'] == 'Africa'),
    (country['continent'] == 'Oceania'),
    (country['continent'] == 'North America'),
    (country['continent'] == 'South America'),
    (country['continent'] == 'Europe')
]
values = [1, 2, 3, 4, 5, 6]
country['continent_code'] = np.select(conditions, values)

continent = country[['continent_code', 'continent']].drop_duplicates()
del country['continent']
```

```python
usRentalDic = {'rentalId':[],
           'rentalDate':[],
           'customerId':[],
           'rt_staffId':[],
           'rt_returnDate':[],
           'rt_lastUpdate':[],
           'inventoryId':[],
           'paymentId':[],
           'py_staffId':[],
           'amount':[],
           'paymentDate':[],
           'py_lastUpdate':[],
           }
start_date = dt.datetime(2010, 1, 1)
end_date = dt.datetime(2020, 12, 31)
rental_id = 400000
payment_id = 121564
min_staff_id = 127
max_staff_id = 186
min_inventory_id = 4582
max_inventory_id = 10473
min_customer_id = 601
max_customer_id = min_customer_id + 600
min_amount = 20
max_amount = 130
for i in range(43567):
    usRentalDic['rentalId'].append(rental_id)
    usRentalDic['rentalDate'].append(fake.date_time_between(start_date=start_date, end_date=end_date))
    usRentalDic['customerId'].append(rn.randint(min_customer_id, max_customer_id))
    usRentalDic['rt_staffId'].append(rn.randint(min_staff_id, max_staff_id))
    usRentalDic['rt_returnDate'].append(usRentalDic['rentalDate'][i]+dt.timedelta(days=rn.randint(1, 9)))
    usRentalDic['rt_lastUpdate'].append(usRentalDic['rt_returnDate'][i]+dt.timedelta(days=1))
    usRentalDic['inventoryId'].append(rn.randint(min_inventory_id, max_inventory_id))
    usRentalDic['paymentId'].append(payment_id)
    usRentalDic['py_staffId'].append(rn.randint(min_staff_id, max_staff_id))
    usRentalDic['amount'].append(rn.randint(min_amount, max_amount))
    usRentalDic['paymentDate'].append(usRentalDic['rentalDate'][i])
    usRentalDic['py_lastUpdate'].append(usRentalDic['paymentDate'][i])
    rental_id += 1
    payment_id += 1
    
usRentalPayment = pd.DataFrame(usRentalDic) 
usRentalPayment
```
